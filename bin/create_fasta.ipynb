# Create fasta from daframe
To be able to split the data aftwerwards

```python
import pandas as pd
from collections import Counter
```

```python
#read data
df = pd.read_csv("../data/final_dataset.csv", index_col=0)
df
```

```python
df['Kingdom'] = df['Kingdom'].str.lower()
df.Kingdom.unique(), len(df)
```

```python
df.columns
```

```python
df['Header'] = '>' + df['Entry'] + '|label=' + df['Cellwall'].astype(str) + df['Extracellular'].astype(str)+ df['Cytoplasmic'].astype(str)+ df['CYtoplasmicMembrane'].astype(str)+ df['OuterMembrane'].astype(str)+ df['Periplasmic'].astype(str)+ df['Flagellar'].astype(str)+ df['Fimbrial'].astype(str) + df['S-layer'].astype(str) + df["Kingdom"]
df.head(2)
```

```python
fasta = dict(zip(df.Header, df.Sequence))
fasta
```

```python
output_path = '../data/fasta_input_graphpart.fasta'
output_file = open(output_path,'w')
for entry, seq in fasta.items():
    # print(type(entry), type(seq))
    identifier_line = entry + "\n"
    output_file.write(identifier_line)
    sequence_line = seq + "\n"
    output_file.write(sequence_line)

output_file.close()
```
